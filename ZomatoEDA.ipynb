```python
df_country = pd.read_excel("Datasets/Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how="left")
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df['Country'].value_counts().index
country_names
```

```python
country_val = final_df['Country'].value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation: Zomato's Maximum records or transactions are from india
After That USA and United Kingdoms

```python
final_df.columns
```

```python
a = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"])
ratings = a.size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

Observaions:
1. When rating is between 4.5 to 4.9 it is excellent
2. When Rating is between 4.0 to 3.4 it is very good
and smililary for good and poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings, palette=['White', 'red','orange','yellow','green','green'])
```

Observations:
1. Not Rated count is very high
2. Maximum No. of ratings are between 2.5 and 3.4

```python
## count plot
sns.countplot(x="Rating color", data = ratings, hue="Rating color", legend=False, palette=['black', 'red','orange','yellow','green','green'])
```
